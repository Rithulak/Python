#Iterator

an Iterator is an object which allows to traverse through all the elements of a collection (list or tuple ) one at a time

implements __iter__() and __next__()

raise stopIteration when no more data is available

used in loops like for loop



```python
class MyNumber:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num<=5:
            x=self.num
            self.num+=1
            return x
        else:
            raise StopIteration

myclass=MyNumber()
myiter=iter(myclass)
for x in myiter:
    print(x)
```

```python
class Square:
    def __iter__(self):
        self.ary=["apple","banana","orange"]
        self.index=0
        return self
    def __next__(self):
        if self.index<len(self.ary):
            val=self.ary[self.index]
            self.index+=1
            return val
        else:
            raise StopIteration
sq=Square()
myiter=iter(sq)
for i in myiter:
    print(i)
```

Generator

A generator is a special type of iterator that is defined using a function and the yield keyword

it is used instead of return
lazy value Generation
memory efficient

```python
def my_gen():
    for i in range(1,20):
        yield i

gen=my_gen()
for i in gen:
    print(i)
```
