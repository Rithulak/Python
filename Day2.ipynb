```python
def get_sum(lst):
    """
    this function returns sum of all elements in list
    """
    sum_=0
    for i in lst:
        sum_+=i
    return sum_

res=get_sum([2,4,5,6,7,8])
print(res)
print(get_sum.__doc__)
#print(sum_) cannot access local variable outside the function
```

```python
class Sports:
    def __init__(self,model,cost,color):
        self.__model=model
        self.__cost=cost
        self.__color=color
    def get_model(self):
        return self.__model
    def set_model(self):
        self.__model=model
    def get_cost(self):
        return self.__cost
    def set_cost(self):
        self.__cost=cost
    def get_color(self):
        return self.__color
    def set_color(self):
        self.__color=color

class Shoes(Sports):
    def __init__(self,model,cost,color,brand):
        super().__init__(model,cost,color)
        self.__brand=brand

Sh=Shoes("A1",23000,"Black","NIke")
print(Sh.get_model())
print(Sh.get_cost())
print(Sh.get_color())
Sh2=Shoes("A2",23000,"White","NIke")
print(Sh2.get_model())
print(Sh2.get_color())
print(Sh2.get_cost())
```

```python
#Keyword Arguments
# it passes all  multiple keword argument into a function (**)
def lst(**kwrgs):
    print(kwrgs)
    if kwrgs:
        print("{0},{1},{2}".format(kwrgs['name'],kwrgs['age'],kwrgs['place']))

lst(name="Rithul",age="21",place="kerala")
```
