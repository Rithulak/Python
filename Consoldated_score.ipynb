```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.csv")
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df[['IDs','Total','Percent']]
```

```python
bins=[0,40,50,60,70,80,90,100]
labels=['<40','<50','<60','<70','<80','<90','<100']
df['group']=pd.cut(df['Percent'],bins=bins,labels=labels,right=False)
range_count=df['group'].value_counts()
range_count
```

```python
plt.bar(labels,range_count,color=['r','g','b','y'])
plt.xlabel("Percentage Group")
plt.ylabel("count")
plt.show()
```
