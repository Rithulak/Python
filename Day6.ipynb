# Scatter plot

```python
x=[10,20,30,40,50]
y=[5,15,25,35,45]

plt.scatter(x,y,marker='o',color='red')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter plot")
plt.show()
```

# BOX Plot
Box plot sumarize the distribution of numerical data by displaying quartiles,outliers, ans potential skewness. Box plot especially use full for identifying outliers

```python
data=[12,15,10,15,17,22,23,25,27,30,35,100,-34]

plt.boxplot(data)

plt.show()
```

```python
quatrs=['Q1','Q2','Q3','Q4']
clothing=[1000,1200,1100,1050]
electronics=[5000,6000,7500,8000]
home_appliance=[3000,7500,8200,900]
plt.figure(figsize=(10,6))
plt.stackplot(quatrs,clothing,electronics,home_appliance,labels=["Clothing","Electronics","Home Appliance"])
plt.xlabel("Quaters")
plt.ylabel("Sales")
plt.title("Product Category sales ovrt sales")
plt.legend()
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
x=np.linspace(-5,-5,100)
y=np.linspace(-5,-5,100)

x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2+y**2))

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=z,cmap='viridis')
ax.set_xlabel("x_axis")
ax.set_ylabel("y_axis")
ax.set_zlabel("Z_axis")
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)

fig,axs=plt.subplots(2,2,figsize=(10,8))

axs[0,0].plot(x,y1,'r')
axs[0,0].set_title("sin wave")
axs[0,1].plot(x,y2,'b')
axs[0,1].set_title("cosin wave")
axs[1,0].plot(x,y3,'g')
axs[1,0].set_title("Tangent Wave")
axs[1,1].plot(x,y4,'m')
axs[1,1].set_title("Exponential Decay")

plt.tight_layout()
plt.show()
```
