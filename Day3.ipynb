```python
def pnl_calc(exit_price,buy_price,qty):
    return (exit_price-buy_price)*qty
def buy_stock(price,qty):
    if price<0 or qty<0:
        return False
    return True
```

```python
import unittest

class Trading(unittest.TestCase):
    def test_pnl(self):
        self.assertEqual(pnl_calc(40,20,40),800)
    def test_buyStock(self):
        self.assertEqual(buy_stock(0,-1),True)

t1=Trading()
t1.test_pnl()
```
