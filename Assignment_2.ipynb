1) why we cannot call the local variable outside the function?

ans) Local variable cannot be accessed outside a function because it is created inside the function's scope and exists only while the function is executing, after which it is destroyed and not visible to the outside of function. Eg:


```python
def sum_():
    a=12
    print(a)
sum_()
```

```python
def sum_():
    a=12
print(a)
sum_()
```

2) where to use Map, Filter and Reduce?

ans) 
Map
    map() is used when the same operation is performed on all elements

```python
num=[1,2,3,4,5]
lst=list(map(lambda x:x**2,num))
print(lst)
```

Filter
    filter() is used when you want to select only specific elements that satisfy a condition 
    

```python
nums=[10,15,20,25,30,35]
even=list(filter(lambda x:x%2==0,nums))
print(even)
```

Reduce
    reduce() is used when you want to perform computation on elements to get single result

```python
from functools import reduce
num=[1,2,3,4]
sum_=reduce(lambda x,y:x+y, num)
print(sum_)
```
